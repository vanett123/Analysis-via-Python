```python
#Word cloud for general questions-major bottlenecks in claims process
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw')
nltk.download('omw-1.4')

# Load the text
text = """
[
••	Lack of database to cheque fraud
•	Lack of expertise in some technical areas
•	Lack of apps and systems that detect some amount of fraud. 
•	Insufficient or Incomplete Documentation from the policyholder
•	Misunderstanding of the Policy coverage
•	Fraudulent Claims
•	Ineffective communication with the policyholder throughout the claim processes
•	Claims involving complex or catastrophic events
]
"""

# Convert to lowercase
text = text.lower()

# Remove punctuation
text = re.sub(r'[^\w\s]', '', text)

# Tokenization
tokens = word_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word not in stop_words]

# Lemmatization
lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(word) for word in tokens]

# Frequency Analysis
word_freq = nltk.FreqDist(tokens)
print("Most common words:", word_freq.most_common(10))

# Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Example text
text = """
•	Structured and Unstructured data 
•	We use external software to analyze the data 
•	Yes, The Driver and Vehicle License Authority Database 
•	Policyholder Information
•	Incident Details
•	Documentation – (photographs, videos, receipts, invoices, repair estimates, police reports, and any other relevant evidence.)
•	Historical Data – (Previous claims filed, the frequency of claims, and any patterns of behavior that may be relevant to the current claim.)
•	Our Risk team Report
•	Structured data is frequently collected using standardized claim forms submitted by policyholders. These forms often include essential details concerning the incident, such as the date, time, location, cause, and scope of the loss or damage. It's normally kept in our internal database. This contains information about policyholders, policies, claims, and finances. Accessing this data includes accessing databases using our technology known as OV3.
•	
•	With Unstructured Data we often collect from various documents such as claim forms, correspondence with policyholders, Assessor or surveyor notes, witness statements, photographs, videos, emails, and other documents.
•	The analysis of structured and unstructured data is an iterative process, where insights gained from initial analyses may lead to further questions and refinements in the analysis approach. Continuous feedback from stakeholders helps ensure that the analysis remains relevant and actionable.
•	Yes, we sought for data externally from our government agencies databases which provide information about the ownership, value, and history of properties involved in claims.
"""

# Topic Modeling
# Step 1: Create a CountVectorizer instance
vectorizer = CountVectorizer(max_df=1.0, min_df=1, stop_words='english')

# Step 2: Transform the text into a document-term matrix
dtm = vectorizer.fit_transform([text])

# Step 3: Create and train the LDA model
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(dtm)

# Step 4: Get the top words for each topic
n_top_words = 5
feature_names = vectorizer.get_feature_names_out()
topic_top_words = []
for topic_idx, topic in enumerate(lda.components_):
    top_words_idx = topic.argsort()[:-n_top_words - 1:-1]
    top_words = [feature_names[i] for i in top_words_idx]
    topic_top_words.append(top_words)

# Step 5: Print the top words for each topic
for topic_idx, words in enumerate(topic_top_words):
    print(f"Topic {topic_idx}: {', '.join(words)}")
```
